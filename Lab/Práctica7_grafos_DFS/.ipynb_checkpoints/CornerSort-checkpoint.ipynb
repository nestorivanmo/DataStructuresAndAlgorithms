```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline
```

```python

contador = 0
listaOrdenada = []

def cornerSort(data,element):
    global contador
    global listaOrdenada

    if len(data) == 1:
        listaOrdenada.append(data[0])
        return
    else:
        low = []
        high = []
        i = 0
        while i < len(data):
            contador += 1
            if data[i] < element:
                low.append(data[i])
            else:
                high.append(data[i])
            i += 1
        if not low:
            if high == data:
                e = high[len(high)-1]
                if high[0] != e:
                    high[0],high[len(high)-1] = high[len(high)-1], high[0]
                    cornerSort(high, e)
                else:
                    x = 0
                    while x < len(high):
                        if data[x] != e:
                            e = data[x]
                            break
                        x += 1
                    if e != high[len(high)-1]:
                        cornerSort(high, e)
                    else:
                        listaOrdenada += high
            else:
                cornerSort(high, high[len(high)-1])
            return
        else:
            cornerSort(low, low[0])
            cornerSort(high, high[len(high)-1])
```

```python
contador = 0
listaOrdenada = []
data = []
for i in range(0,100):
    data.append(random.randint(0,100))

print("Unsorted list : " + str(len(data)) + "\n")
print(data)
print("\n\nNow sorting...\n")
cornerSort(data,data[0])
print("\nSorted list : " + str(len(listaOrdenada)) + "\n")
print(listaOrdenada)
print("\n"+str(len(data)) + " elements : " + str(contador) + " iterations")
```

```python
x = []
y = []
length = 10
for i in range(1,length):
    x.append(i)
    d = []
    for j in range(i):
        d.append(random.randint(0,500))
    contador = 0
    cornerSort(d,d[0])
    y.append(contador)
        
plt.plot(x, y, 'g')
plt.show()
```
