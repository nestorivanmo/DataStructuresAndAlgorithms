```python
import random
import matplotlib.pyplot as plt
```

```python
key = ['','+','*','^', 'A', 'B','/', '3', '','','','', 'C', 'D']
orderCounter = 0
preOrderCounter = 0
posOrderCounter = 0
```

```python
def left(x):
    return 2*x

def right(x):
    return (2*x) + 1

def orderedTreeWalk(x, tree):
    global orderCounter
    if x < len(tree):
        orderCounter += 1
        orderedTreeWalk(left(x), tree)
        #print(tree[x])
        orderedTreeWalk(right(x),tree)

def preOrderedTreeWalk(x, tree):
    global preOrderCounter
    if x < len(tree):
        preOrderCounter += 1
        #print(tree[x])
        preOrderedTreeWalk(left(x),tree)
        preOrderedTreeWalk(right(x),tree)

def postOrderedTreeWalk(x, tree):
    global posOrderCounter
    if x < len(tree):
        posOrderCounter += 1
        postOrderedTreeWalk(left(x),tree)
        postOrderedTreeWalk(right(x),tree)
        #print(tree[x])
```

```python
print("Inorder Tree Walk")
orderedTreeWalk(1, key)
print("Preorder Tree Walk")
preOrderedTreeWalk(1, key)
print("Postorder Tree Walk")
postOrderedTreeWalk(1, key)
```

```python
#Gr√°ficas 

x = []
orderedY = []

for i in range(1,2000):
    x.append(i)
    tree = ['']
    orderCounter = 0
    for j in range(1,i):
        tree.append(j)
    orderedTreeWalk(1, tree)
    orderedY.append(orderCounter)
    
plt.plot(x,orderedY,color="r")
plt.title = "Inordered Tree Walk"
```

```python
x = []
preOrderedY = []

for i in range(1,2000):
    x.append(i)
    tree = ['']
    preOrderCounter = 0
    for j in range(1,i):
        tree.append(j)
    preOrderedTreeWalk(1, tree)
    preOrderedY.append(preOrderCounter)
    
plt.plot(x,orderedY,color="g")
plt.title = "Preordered Tree Walk"
```

```python
x = []
postOrderedY = []

for i in range(1,2000):
    x.append(i)
    tree = ['']
    posOrderCounter = 0
    for j in range(1,i):
        tree.append(j)
    postOrderedTreeWalk(1, tree)
    postOrderedY.append(posOrderCounter)
    
plt.plot(x,postOrderedY)
plt.title = "Postordered Tree Walk"
```
