```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline
import math
from Node import *
```

```python
TAM = 1000
counter = 0
```

```python
def hashNestor(number):
    position = ""
    number = str(number)
    lista = list(number)
    position += number[0]
    position += number[len(number)//2]
    position += number[len(number)-1]
    return int(position)
```

```python
def generateHashTable():
    return [[None] * TAM for _ in range(TAM)]

def insertIntoHashTable(listOfItems, arrayIn2D):
    for k in range(TAM):
        key = hashNestor(listOfItems[k].id)
        if arrayIn2D[key][0] is None:
            arrayIn2D[key][0] = listOfItems[k]
        else:
            positionToInsert = 0
            while(arrayIn2D[key][positionToInsert] is not None):
                positionToInsert += 1
            arrayIn2D[key][positionToInsert] = listOfItems[k]
            

def searchInHashTable(element, arrayIn2D):
    global counter
    key = hashNestor(element)
    if arrayIn2D[key][0] is None:
        counter = 1
        return "City not found"
    else:
        counter = 0
        positionToLookAt = 0
        while (arrayIn2D[key][positionToLookAt] is not None):
            counter += 1
            if arrayIn2D[key][positionToLookAt].id == element:
                return "City found : " + str(arrayIn2D[key][positionToLookAt].city)
            positionToLookAt += 1  
           
        
    
def printHashTable(arrayIn2D):
    total = 0
    for i in range(TAM):
        if arrayIn2D[i][0] is not None:
            print(i)
            for j in range(TAM):
                if arrayIn2D[i][j] is not None:
                    total += 1
                    print(arrayIn2D[i][j].city +" -> " + str(arrayIn2D[i][j].id))
```

```python
hashTable = generateHashTable()
cities = readFromFile()
insertIntoHashTable(cities, hashTable)
printHashTable(hashTable)
```

```python
print("Best Case")
x = []
y = []
counter = 0
for i in range(TAM):
    if hashTable[i][0] is not None:
        x.append(i)
        counter = 0
        searchInHashTable(hashTable[i][0].id,hashTable)
        y.append(counter)
            
print(len(x))
print(len(y))
plt.plot(x, y, 'g')
plt.show()
```

```python
print("Average Case")
x = []
y = []
counter = 0
for i in range(TAM):
    if hashTable[i][0] is not None:
        x.append(i)
        pos = 0
        for j in range(TAM):
            if hashTable[i][j] is not None:
                pos += 1
        randomPos = random.randint(0, pos-1)
        counter = 0
        searchInHashTable(hashTable[i][randomPos].id, hashTable)
        y.append(counter)
            
            
print(len(x))
print(len(y))
plt.plot(x, y)
plt.show()
```

```python
print("Worst Case")
x = []
y = []
counter = 0
for i in range(TAM):
    if hashTable[i][0] is not None:
        x.append(i)
        pos = 0
        for j in range(TAM):
            if hashTable[i][j] is not None:
                pos += 1
        counter = 0
        searchInHashTable(hashTable[i][pos-1].id, hashTable)
        y.append(counter)
            
            
print(len(x))
print(len(y))
plt.plot(x, y, 'r')
plt.show()
```
