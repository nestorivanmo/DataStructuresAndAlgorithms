```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline
```

```python
def particionar(A,low,high):
	pivote = A[high]
	i = ( low - 1 )
	for j in range(low, high):
		if A[j] <= pivote:
			i += 1
			A[i], A[j] = A[j], A[i]
	A[i+1], A[high] = A[high], A[i+1]
	return (i + 1)

def quickSort(A, p, r):
	print(p,r)
	if p < r:
		q = particionar(A, p, r)
		print("izq:",p, q-1)
		quickSort(A, p, q-1)
		print("der:",q+1,r)
		quickSort(A, q+1, r)
	return A
```

```python
def probarQuickSort(elementos):
	lista = []
	for x in xrange(0,elementos):
		lista.append(random.randint(0,elementos*2))
	print(quickSort(lista, 0,len(lista)-1))	
```

```python
print("\nLista con 10 elementos")
probarQuickSort(10)
print("\n")

print("\nLista con 50 elementos")
probarQuickSort(50)
print("\n")

print("\nLista con 100 elementos")
probarQuickSort(100)
print("\n")
```

```python
contador = 0

def particionarGraf(A,low,high):
	global contador
	pivote = A[high]
	i = ( low - 1 )
	for j in range(low, high):
		contador += 1
		if A[j] <= pivote:
			i += 1
			A[i], A[j] = A[j], A[i]
	A[i+1], A[high] = A[high], A[i+1]
	return (i + 1)

def quickSortGraf(A, p, r):
	global contador
	if p < r:
		contador += 1
		q = particionarGraf(A, p, r)
		quickSortGraf(A, p, q-1)
		quickSortGraf(A, q+1, r)
	return contador
```

```python
def graficarPeor(elementos):
	global contador
	lista = []
	x = []
	yprom = []
	ymejor = []
	ypeor = []
	for i in xrange(1, elementos+1):
		x.append(i)
		lista=[]
		for j in xrange(1, i+1):
			lista.append(random.randint(-elementos*2,elementos*2))
		contador = 0
		lista.reverse()
		ypeor.append(quickSortGraf(lista,0,len(lista)-1))
	return (x, ypeor)
```

```python
def graficarPromedio(elementos):
	global contador
	lista = []
	x = []
	yprom = []
	ymejor = []
	ypeor = []
	for i in xrange(1, elementos+1):
		x.append(i)
		lista=[]
		for j in xrange(1, i+1):
			lista.append(random.randint(-elementos*2,elementos*2))
		contador = 0
		yprom.append(quickSortGraf(lista, 0, len(lista)-1))
		contador = 0
	return (x, yprom)
```

```python
def graficarMejor(elementos):
	global contador
	lista = []
	x = []
	yprom = []
	ymejor = []
	ypeor = []
	for i in xrange(1, elementos+1):
		x.append(i)
		lista=[]
		for j in xrange(1, i+1):
			lista.append(random.randint(-elementos*2,elementos*2))
		contador = 0
		yprom.append(quickSortGraf(lista, 0, len(lista)-1))
		contador = 0
	return (x, yprom)
```
